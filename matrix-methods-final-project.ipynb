```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/kaggle/input/criscpr/CRISPRGeneEffect.csv') 

```

## Load the dataset form the CSV file

```python
data.head()
```

```python
data = data.drop(data.columns[0], axis=1)
data.head()
```

## Some preprocessing(data looks pretty good already)

```python
missing_values_count = data.isna().sum()

# Display the count of missing values for each column
print("Number of missing values in each column:")
print(missing_values_count)
```

## Scaling

```python
from sklearn.impute import SimpleImputer 

#All NAN => Mean
imputer = SimpleImputer(strategy='mean') 
data_imputed = pd.DataFrame(imputer.fit_transform(data))


scaler = StandardScaler()
data_scaled = pd.DataFrame(scaler.fit_transform(data_imputed), columns = data.columns)

##Idk if you want pca but ye
pca = PCA(n_components=10)  
data_pca = pca.fit_transform(data_scaled)

data_pca_df = pd.DataFrame(data_pca, columns=[f'PC{i+1}' for i in range(10)])  # Naming the columns as PC1, PC2, ..., PC10


explained_variance_ratio = pca.explained_variance_ratio_

plt.figure(figsize=(10, 6))
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, alpha=0.5, align='center')
plt.title('Explained Variance Ratio')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.xticks(range(1, len(explained_variance_ratio) + 1))
plt.grid(True)
plt.show()
```
